```python
import csv

# User Defined Functions

def display():
  try:
     with open("/content/drive/MyDrive/colleges.csv","r") as cF:
      cR = csv.reader(cF)
      for r in cR:
        print(r)
  except FileNotFoundError:
    print("colleges.csv is not available.")

def Adm_addnewuni():
  R=[]
  with open("/content/drive/MyDrive/colleges.csv","a",newline="") as F:
    CW = csv.writer(F)
    while True:
      Name=(input("Name of Univerity: ")).title()
      control=(input("Is the University Public/Private?: ")).title()
      region=(input("Region of USA: ")).title()
      locale=(input("Locale of the University: ")).title()
      admrate=float(input("Admission Rate: "))
      ACT=int(input("ACT Median of University: "))
      SAT=int(input("SAT Median of University: "))
      parttime=float(input("Percentage of part-time undergrads: "))
      avgcost=int(input("Average cost of the University: "))
      expstu=int(input("University expenditure per student: "))
      retentionrate=float(input("Retention rate of first time students: "))
      avgincome=int(input("Average family income mean earnings 8 years after entry at the University: "))

      R.append([Name, control, region, locale, admrate, ACT, SAT, parttime, avgcost, expstu, retentionrate, avgincome])

      Ch=input("Do you want to add another university?(yes/no)): ")
      if Ch.lower()=="no":
        break
    print("Record Added: ",R)
    CW.writerows(R)

def Adm_addnewattribute():
    newname=input("Name of new attribute: ")
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
    with open("/content/drive/MyDrive/colleges.csv","w",newline="") as F:
      CW = csv.writer(F)
      header = CR[0]
      header.append(newname)
      CW.writerow(header)
      for i in range(1,len(CR)):
        univdata = CR[i]
        newattrdata = input("New attribute value for " + univdata[0] + ": ")
        univdata.append(newattrdata)
        CW.writerow(univdata)
  
def Adm_modify():
    uniname=(input("Name of University that is to be modified: ")).title()
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
    header=CR[0]
    print("Attributes present in the file: ", header)
    attrname=input("Name of Attribute that is to be modified: ")
    attrval=input("Value of Modified Attribute: ")
    try:
      attrind=header.index(attrname)
      with open("/content/drive/MyDrive/colleges.csv","w",newline="") as F:
        CW = csv.writer(F)
        CW.writerow(CR[0])
        for i in CR[1:]:
          if i[0]==uniname:
            i[attrind]=attrval
            CW.writerow(i)
          else:
            CW.writerow(i)
    except:
      print("Incorrect Atrribute requested")

def Adm_delete():
    uniname=(input("Name of University that is to be deleted: ")).title()
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
    with open("/content/drive/MyDrive/colleges.csv","w",newline="") as F:
      CW = csv.writer(F)
      CW.writerow(CR[0])
      for i in CR[1:]:
        if i[0]==uniname:
          continue
        else:
          CW.writerow(i)
  
reviewunisearch=[]
def user_searchuni():
  with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
      while True:
        uniname=(input("Enter the name of the university: ")).title()
        for i in CR:
          if i[0]==uniname:
            print(CR[0])
            print(i)
            global reviewunisearch
            reviewunisearch.append(i)
            break
        else:
          print("This University is not in our database.")
          a=input("Do you want this University to be added?(Yes/No): ")
          if a.lower()=="yes":
            user_requestnewuni()
        choice=input("Do you want to search another University?(Yes/No): ")
        if choice.lower()=="no":
          break

def user_sort():
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
    header=CR[0]
    print("Attributes present in the file: ", header)
    attrname=input("Name of Attribute on which the Universities have to sorted: ")
    try:
      attrind=header.index(attrname)
      CRcopy=CR[1:]
      N=len(CRcopy) 
      for i in range(N-1):   
        for j in range(N-i-1): 
          if CRcopy[j][attrind] > CRcopy[j+1][attrind]:   
            CRcopy[j],CRcopy[j+1]=CRcopy[j+1],CRcopy[j] 
      print("Universities sorted based on "+attrname+": ")
      print(CR[0])
      for i in CRcopy:
        print(i)
    except:
      print("Incorrect Atrribute requested")

requestnewuni=[]
def user_requestnewuni():
  global requestnewuni
  with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
      while True:
        newuni=input("Enter the name of the university whose data should be added in the database: ")
        for i in CR:
          if i[0]==newuni:
            print("University is already in database.")
            break
        else:
          requestnewuni.append(newuni)
        choice=input("Do you want to add another?(Yes/No): ")
        if choice.lower()=="no":
          break

def user_selective():
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR=list(csv.reader(F))
    header=CR[0]
    print("Attributes present in the file: ", header)
    attrname=input("Name of Attribute to filter selectively: ")
    try:
      attrind=header.index(attrname)
      print((header[0],attrname))
      for i in CR:
        print((i[0],i[attrind]))
    except:
      print("Incorrect Attribute requested")

def sumtestscores():
  try:
    with open("/content/drive/MyDrive/colleges.csv","r") as F:
      CR = list(csv.reader(F, delimiter=','))
      s_public_ACT=0
      s_public_SAT=0
      s_private_ACT=0
      s_private_SAT=0
      count_public=0
      count_private=0
      for i in range(1,len(CR)):
        if CR[i][1]=="Public":
          s_public_ACT+=int(CR[i][5])
          s_public_SAT+=int(CR[i][6])
          count_public+=1
        else:
          s_private_ACT+=int(CR[i][5])
          s_private_SAT+=int(CR[i][6])
          count_private+=1
    return s_public_ACT, s_public_SAT, s_private_ACT, s_private_SAT, count_public, count_private
  except:
    print("colleges.csv is not available") 


def avgpublic_ACT():
  tvalues=sumtestscores()
  print("Average ACT Scores Of Public Universities:", tvalues[0]//tvalues[4])

def avgpublic_SAT():
  tvalues=sumtestscores()
  print("Average SAT Scores Of Public Universities:", tvalues[1]//tvalues[4])
  
def avgprivate_ACT():
  tvalues=sumtestscores()
  print("Average ACT Scores Of Private Universities:", tvalues[2]//tvalues[5])

def avgprivate_SAT():
  tvalues=sumtestscores()
  print("Average SAT Scores Of Private Universities:", tvalues[3]//tvalues[5])

# Main Code

print("Welcome to UniInfo Management System!")
print()
while True:
  choice=input("For Signing in press S\nTo see SAT\\ACT Averages press R\nTo exit press E -  ")
  print()
  if choice.lower()=="s":
    c=input("Admin: A\nUser: U - ")
    if c.lower()=="a":
      id=input("Username: ")
      password=input("Password: ")
      with open("/content/drive/MyDrive/admin.csv","r") as F:
         Ad=list(csv.reader(F))
         for i in Ad:
           if i[1]==id:
             break
         else:
           with open("/content/drive/MyDrive/admin.csv","a",newline="") as R:
              Adcw = csv.writer(R)
              print("To create your new Admin account, enter the following details")
              n=(input("Name: ")).title()
              username=(input("Username: "))
              passw=(input("Password: "))
              Adcw.writerow([n,username,passw])
              print()
              print("Your accoount has been created.")
      if  requestnewuni==[]:
        print("No request for addition of a new university")
      else:
        print("Univeristy addition Requests")
        for i in requestnewuni:
          print(i)
        yn=input("Do you want to add these universities (Yes/No)): ")
        if yn.lower()=="yes":
          Adm_addnewuni()
      while True:
        print()
        function=input("Add New University Record:A\nModify Existing University Record:M\nShow Existing Records:S\nDelete University Record:D\nAdd New Attribute:N\nQuit:Q - ")
        f=function.lower()
        print()
        if f=="a":
          Adm_addnewuni()
        elif f=="n":
          Adm_addnewattribute()
        elif f=="s":
          display()
        elif f=="m":
          Adm_modify()
        elif f=="d":
          Adm_delete()
        elif f=="q":
          break
    elif c.lower()=="u":
      id=input("Username: ")
      password=input("Password: ")
      with open("/content/drive/MyDrive/user.csv","r") as F:
         Us=list(csv.reader(F))
         for i in Us:
           if i[1]==id:
             break
         else:
             with open("/content/drive/MyDrive/user.csv","a",newline="") as R:
                uscw = csv.writer(R)
                print("To create your new account, enter the following details")
                n=(input("Name: ")).title()
                username=(input("Username: "))
                passw=(input("password: "))
                uscw.writerow([n,username,passw])
                print()
                print("Your accoount has been created.")
      while True:
        print()
        function=input("Display:D\nSorting:S\nSearch a University Record:C\nRequest Addition of University:A\nReview University Search:R\nSelective Filtering of Universities:F\nQuit:Q - ")
        f=function.lower()
        print()
        if f=="a":
          user_requestnewuni()
        elif f=="c":
          user_searchuni()
        elif f=="r":
          print("You searched these universites: ")
          for i in reviewunisearch:
            print(i)
        elif f=="s":
          user_sort()
        elif f=="d":
          display()
        elif f=="f":
          user_selective()
        elif f=="q":
          break
  elif choice.lower()=="r":
    while True:
      print()
      print("Check Test Score Report")
      function=input("Average ACT Score of Public Universities:A\nAverage SAT Score of Public Universities:B\nAverage ACT Score of Private Universities:C\nAverage SAT Score of Private Universities:D\nQuit:Q - ")
      f=function.lower()
      print()
      if f=="a":
        avgpublic_ACT()
      elif f=="b":
        avgpublic_SAT()
      elif f=="c":
        avgprivate_ACT()
      elif f=="d":
        avgprivate_SAT()
      elif f=="q":
        break
      print()
  else:
    print("Thankyou!")
    break
```
